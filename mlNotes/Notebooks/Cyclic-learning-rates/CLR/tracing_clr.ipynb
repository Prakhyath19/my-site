```python
import math
import matplotlib.pyplot as plt
```

```python
#variables
base_lr = 0.001
max_lr = 0.006
step_size = 2000
lr_list = []
itr_list = []
```

```python
def traingular_clr(iteration, step_size, base_lr, max_lr):
    cycle = math.floor(1+iteration/(2*step_size))
    x = abs(iteration/step_size - (2*cycle)+1)
    lr = base_lr + ((max_lr-base_lr) * max(0,(1-x)))
    return lr
```

```python
def traingular2_clr(iteration, step_size, base_lr, max_lr):
    cycle = math.floor(1+iteration/(2*step_size))
    x = abs(iteration/step_size - (2*cycle)+1)
    lr = base_lr + ((max_lr-base_lr) * max(0,(1-x))) / float(2**(cycle-1))
    return lr
```

```python
#plotting the model
def plotModel(policy, itr_list, lr_list):
    # plt.figure(figsize=(12,8))
    fig = plt.figure()
    plt.xlabel("iteration")
    plt.ylabel("Learning Rate")
    plt.title(policy)
    plt.plot(itr_list, lr_list)
    # plt.show()                                                                                                        
    plt.clf(fig)
```

```python
for iteration in range(10000):
    lr = traingular_clr(iteration, step_size, base_lr,max_lr)
    lr_list.append(lr)
    itr_list.append(iteration)
    # print(f"Iteration: {iteration}, Learning Rate: {lr: .6f}")
plotModel("traingular_clr", itr_list, lr_list)
```

```python
for iteration in range(10000):
    lr = traingular2_clr(iteration, step_size, base_lr,max_lr)
    lr_list.append(lr)
    itr_list.append(iteration)
    # print(f"Iteration: {iteration}, Learning Rate: {lr: .6f}")
plotModel("traingular2_clr", itr_list, lr_list)
```
